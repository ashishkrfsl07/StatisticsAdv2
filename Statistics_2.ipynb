# Q1. What is hypothesis testing in statistics?


Hypothesis testing is a formal procedure in statistical inference that uses data from a sample to draw conclusions about a population parameter or a population probability distribution. Essentially, it's a structured method for deciding whether the evidence from your sample is strong enough to reject a particular claim about the population.

This process allows researchers and analysts to quantify the likelihood that an observed effect or relationship in the sample data occurred by chance, as opposed to reflecting a true effect in the larger population.

Key Components

Hypothesis testing is built around two mutually exclusive statements about the population:

    Null Hypothesis (H0​): This is the "default" assumption, typically stating that there is no effect, no difference, or no relationship between variables. It represents the status quo. For example: "The average height of men and women is the same (μ1​=μ2​)."

Alternative Hypothesis (Ha​ or H1​): This is the research claim or what you are trying to find evidence for, stating that there is an effect, difference, or relationship. It's the opposite of the null hypothesis. For example: "The average height of men and women is different (μ1​=μ2​)."

Steps in the Procedure

The process generally follows these steps:

    State the Hypotheses: Define the null (H0​) and alternative (Ha​) hypotheses.

Choose the Significance Level (α): This is the maximum allowable probability of making a Type I Error (rejecting a true H0​). Common choices are α=0.05 (5%) or α=0.01 (1%).

Collect Data and Calculate the Test Statistic: Gather a representative sample and compute a test statistic (e.g., z-score, t-score) that measures how far the sample results deviate from what's expected under the null hypothesis.

Determine the p-value: The p-value is the probability of observing the sample data (or data more extreme) if the null hypothesis (H0​) were actually true.

    Make a Decision: Compare the p-value to the significance level (α):

    If p-value≤α, you Reject the Null Hypothesis—the sample evidence is strong enough to conclude the alternative hypothesis is likely true.

If p-value>α, you Fail to Reject the Null Hypothesis—the sample evidence is not strong enough to conclude the alternative hypothesis is true. Note: Failing to reject H0​ does not prove it is true; it just means there isn't sufficient evidence to reject it.

Draw a Conclusion: State the final findings in the context of the original research question.

Potential Errors

Since hypothesis testing relies on samples, there's always a possibility of error:

    Type I Error (False Positive): Rejecting the null hypothesis when it is actually true. The probability of this error is set by the significance level (α).

Type II Error (False Negative): Failing to reject the null hypothesis when it is actually false. The probability of this error is denoted by β.

# Q3.Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.


The significance level, denoted by the Greek letter α (alpha), is a critical threshold in hypothesis testing that determines how strong the evidence must be to reject the null hypothesis (H0​).

Significance Level (α)

The significance level is formally defined as the maximum probability of making a Type I Error (false positive), which is the error of rejecting the null hypothesis when it is actually true.

    It is a value set by the researcher before the test is conducted.

The most common value for α in many fields is 0.05 (or 5%), meaning the researcher is willing to accept a 5% chance of incorrectly concluding that an effect exists when it does not. Other common values are 0.01 or 0.10.

Role in Deciding the Outcome

The significance level is the yardstick used to compare against the calculated p-value from your data to make a final decision about the null hypothesis.

The Decision Rule

    Calculate the p-value: The p-value is the probability of observing your sample data (or data more extreme) if the null hypothesis were true. It measures the strength of evidence against H0​.

Compare p-value to α:

    If p≤α: The results are considered statistically significant. The evidence against H0​ is strong enough (i.e., the observed data is very unlikely under H0​), so you reject the null hypothesis in favor of the alternative hypothesis (Ha​).

If p>α: The results are not statistically significant. The evidence is not strong enough, so you fail to reject the null hypothesis.

Impact of α on the Test

The choice of α directly controls the trade-off between the two types of errors:

    Lower α (e.g., 0.01): This is a stricter standard. It reduces the risk of a Type I Error (false positive) but increases the risk of a Type II Error (false negative: failing to detect a real effect). You need very strong evidence (a very small p-value) to reject H0​.

Higher α (e.g., 0.10): This is a more lenient standard. It increases the risk of a Type I Error but decreases the risk of a Type II Error. You can reject H0​ with weaker evidence.

# Q4. What are Type I and Type II errors? Give examples of each.


In statistical hypothesis testing, Type I and Type II errors are the two potential incorrect conclusions a researcher can make when using sample data to draw inferences about a population.

Type I Error (α) - The False Positive

A Type I error occurs when you reject the null hypothesis (H0​) when it is actually true.

    Also known as: A False Positive (a "false alarm").

Probability: The probability of a Type I error is denoted by α (alpha), which is the significance level set by the researcher before the test.

Definition: Concluding that an effect, difference, or relationship exists when, in reality, it does not.

Example of a Type I Error: Medical Testing

    Scenario: Testing a healthy patient for a rare disease.

Null Hypothesis (H0​): The patient does not have the disease.

Type I Error: The test result is positive (reject H0​), but the patient is actually healthy (H0​ is true).

    Consequence: The patient is subjected to unnecessary and potentially harmful treatment, stress, and expense.

Type II Error (β) - The False Negative

A Type II error occurs when you fail to reject the null hypothesis (H0​) when it is actually false.

    Also known as: A False Negative (a "miss").

Probability: The probability of a Type II error is denoted by β (beta), which is inversely related to the test's power (1−β).

Definition: Failing to conclude that an effect, difference, or relationship exists when, in reality, it truly does.

Example of a Type II Error: Clinical Trial

    Scenario: Testing a new drug that is genuinely effective.

    Null Hypothesis (H0​): The new drug has no effect (it is not better than the old treatment/placebo).

Type II Error: The study concludes the drug has no effect (fail to reject H0​), but the drug is actually effective (H0​ is false).

    Consequence: A genuinely effective new drug is mistakenly abandoned, and patients are denied a superior treatment option.

## Q5. What is the difference between a Z-test and a T-test? Explain when to use each.


The fundamental difference between a Z-test and a T-test lies in the information you have about the population standard deviation (σ) and the sample size (n). Both are parametric statistical tests used to determine if the mean of a sample (or the difference between two means) is significantly different from a hypothesized value or another mean. The Z-test is used when the population standard deviation (σ) is known or when the sample size (n) is large (typically n≥30), which allows the sample's standard deviation to reliably approximate the population's, and the test statistic follows the standard normal (Z) distribution. In contrast, the T-test is used when the population standard deviation (σ) is unknown and the sample size (n) is small (typically n<30); in this case, the sample standard deviation is used as an estimate, which introduces more uncertainty, so the test statistic follows the Student's t-distribution, which has heavier tails than the Z-distribution to account for this extra variability. As the sample size increases (approaching n=30), the t-distribution becomes virtually identical to the Z-distribution.

# Q6.  Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.


```python
import numpy as np
import matplotlib.pyplot as plt

# --- 1. Define Binomial Parameters ---
# Number of trials (flips/attempts) in one experiment
n = 10
# Probability of success in a single trial (e.g., probability of heads)
p = 0.5
# Number of times to run the experiment (sample size for the plot)
sample_size = 10000

# --- 2. Generate Binomial Distribution Data ---
# Generate 'sample_size' random numbers, where each number is the sum of 'n' trials
# with success probability 'p'.
data = np.random.binomial(n=n, p=p, size=sample_size)

# --- 3. Plot the Histogram ---
plt.figure(figsize=(10, 6))

# Define bins centered on the integer values (0 to n)
# np.arange(n + 2) creates [0, 1, 2, ..., 11]
# Subtracting 0.5 creates bins: [-0.5, 0.5, 1.5, ..., 10.5]
bins = np.arange(n + 2) - 0.5

# Plot the histogram
plt.hist(data, bins=bins, rwidth=0.8, color='#3498db', edgecolor='black', alpha=0.9)

# --- 4. Add Labels and Title for Clarity ---
plt.title(f'Histogram of Binomial Distribution (n={n}, p={p}) - {sample_size} Samples', fontsize=16)
plt.xlabel('Number of Successes (k)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

# Set x-axis ticks to show only integer values (0, 1, ..., 10)
plt.xticks(np.arange(n + 1))
plt.grid(axis='y', alpha=0.5)

# Show the plot
plt.show()

# --- 5. Output Summary (Optional but helpful) ---
print(f"Generated {sample_size} samples from Binomial(n={n}, p={p}).")
print(f"The theoretical mean is n * p = {n * p}.")
print(f"The sample mean is: {np.mean(data):.4f}.")
```

# Q7. Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results. sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6, 50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5, 50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9, 50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

```python
import numpy as np
from scipy.stats import norm

# --- 1. Data and Assumptions ---

# The sample dataset provided by the user (n=36)
sample_data = [
    49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
    50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
    50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
    50.3, 50.4, 50.0, 49.7, 50.5, 49.9
]

# Z-Test requires assumed population parameters:
# Hypothesized Population Mean (μ₀)
mu_0 = 50.0
# Known Population Standard Deviation (σ) - This must be known for a Z-test.
sigma = 1.2
# Significance Level (alpha)
alpha = 0.05

# --- 2. Calculate Sample Statistics ---

sample_mean = np.mean(sample_data)
sample_size = len(sample_data)
standard_error = sigma / np.sqrt(sample_size)

# --- 3. Hypothesis Formulation ---
print("--- Z-Test Hypothesis Test ---")
print(f"Sample Size (n): {sample_size}")
print(f"Sample Mean (x̄): {sample_mean:.4f}")
print(f"Hypothesized Population Mean (μ₀): {mu_0}")
print(f"Known Population Standard Deviation (σ): {sigma}")
print(f"Significance Level (α): {alpha}\n")

print(f"Null Hypothesis (H₀): The true population mean is equal to {mu_0}.")
print(f"Alternative Hypothesis (Hₐ): The true population mean is NOT equal to {mu_0} (Two-tailed test).\n")

# --- 4. Calculate Z-Statistic ---

# Z = (Sample Mean - Hypothesized Mean) / Standard Error
z_statistic = (sample_mean - mu_0) / standard_error

print(f"Calculated Z-Statistic: {z_statistic:.4f}")

# --- 5. Calculate P-value (Two-tailed) ---

# The p-value for a two-tailed test is 2 * P(Z > |z_statistic|)
# Use the survival function (sf) for P(Z > z)
p_value = 2 * norm.sf(abs(z_statistic))

print(f"Calculated P-value: {p_value:.4f}")

# --- 6. Decision and Interpretation ---

print("\n--- Decision ---")
if p_value < alpha:
    print(f"Since the P-value ({p_value:.4f}) is less than alpha ({alpha}), we REJECT the Null Hypothesis.")
    print(f"Conclusion: There is statistically significant evidence that the true population mean is NOT equal to {mu_0}.")
else:
    print(f"Since the P-value ({p_value:.4f}) is greater than alpha ({alpha}), we FAIL TO REJECT the Null Hypothesis.")
    print(f"Conclusion: There is NOT enough statistically significant evidence to conclude that the true population mean is different from {mu_0}.")

print("----------------------------")
```

# Q8. Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t, norm

# --- 1. Simulation Parameters ---

# Parameters used for generating the theoretical normal distribution
POP_MEAN = 75.0
POP_STD = 8.0
SAMPLE_SIZE = 100
CONFIDENCE_LEVEL = 0.95
alpha = 1 - CONFIDENCE_LEVEL

# --- 2. Generate Sample Data ---

# Generate data from a normal distribution
np.random.seed(42) # for reproducibility
sample_data = np.random.normal(loc=POP_MEAN, scale=POP_STD, size=SAMPLE_SIZE)

# --- 3. Calculate Sample Statistics and T-Interval (Standard Practice for CI) ---

sample_mean = np.mean(sample_data)
# Use ddof=1 for sample standard deviation (unbiased estimator)
sample_std = np.std(sample_data, ddof=1)
df = SAMPLE_SIZE - 1 # Degrees of Freedom

# Calculate Standard Error (SE)
standard_error = sample_std / np.sqrt(SAMPLE_SIZE)

# Calculate the critical T-value for a two-tailed test
# ppf is the percent point function (inverse of CDF)
t_critical = t.ppf(1 - alpha/2, df)

# Calculate Margin of Error (ME)
margin_of_error = t_critical * standard_error

# Calculate Confidence Interval (CI = Sample Mean ± ME)
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# --- 4. Print Results ---
print("--- Confidence Interval Calculation ---")
print(f"Sample Size (n): {SAMPLE_SIZE}")
print(f"Sample Mean (x̄): {sample_mean:.4f}")
print(f"Sample Standard Deviation (s): {sample_std:.4f}")
print(f"Degrees of Freedom (df): {df}")
print(f"Confidence Level: {CONFIDENCE_LEVEL*100}%")
print(f"Critical T-value: {t_critical:.4f}")
print(f"Margin of Error: {margin_of_error:.4f}")
print(f"95% Confidence Interval: ({ci_lower:.4f}, {ci_upper:.4f})")
print(f"Interpretation: We are 95% confident that the true population mean lies between {ci_lower:.4f} and {ci_upper:.4f}.")
print("---------------------------------------\n")


# --- 5. Plotting the Data and CI ---

plt.figure(figsize=(12, 7))

# Plot the histogram of the sampled data
plt.hist(sample_data, bins=15, density=True, color='#2ecc71', edgecolor='black', alpha=0.7, label='Sample Data Histogram')

# Plot a fitted normal distribution curve for visual comparison
x = np.linspace(sample_data.min(), sample_data.max(), 100)
# Use the sample mean and sample standard deviation to fit the curve
p = norm.pdf(x, sample_mean, sample_std)
plt.plot(x, p, 'k', linewidth=2, label='Fitted Normal PDF')

# Add Mean Line
plt.axvline(sample_mean, color='red', linestyle='--', linewidth=2, label=f'Sample Mean ({sample_mean:.2f})')

# Add Confidence Interval Lines
# Lower CI bound
plt.axvline(ci_lower, color='blue', linestyle='-', linewidth=3, label=f'{CONFIDENCE_LEVEL*100}% CI Bound')
# Upper CI bound
plt.axvline(ci_upper, color='blue', linestyle='-', linewidth=3)
# Shade the Confidence Interval region
plt.axvspan(ci_lower, ci_upper, color='blue', alpha=0.1, label='Confidence Interval Region')


# Add Plot Labels and Title
plt.title(f'Sampled Data and {CONFIDENCE_LEVEL*100}% Confidence Interval for the Mean', fontsize=16)
plt.xlabel('Data Values', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.legend()
plt.grid(axis='y', alpha=0.4)

plt.show()
```

# Q9. Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t, norm

# --- Function to Calculate Z-Scores ---
def calculate_z_scores(data):
    """Calculates the Z-score for each data point in the dataset (Standardization).

    Z-score = (x - sample_mean) / sample_std
    """
    mean = np.mean(data)
    # Use ddof=1 for sample standard deviation (unbiased estimator)
    std_dev = np.std(data, ddof=1)

    # Avoid division by zero if all data points are identical
    if std_dev == 0:
        return np.zeros_like(data)

    z_scores = (data - mean) / std_dev
    return z_scores

# --- 1. Simulation Parameters ---

# Parameters used for generating the theoretical normal distribution
POP_MEAN = 75.0
POP_STD = 8.0
SAMPLE_SIZE = 100
CONFIDENCE_LEVEL = 0.95
alpha = 1 - CONFIDENCE_LEVEL

# --- 2. Generate Sample Data ---

# Generate data from a normal distribution
np.random.seed(42) # for reproducibility
sample_data = np.random.normal(loc=POP_MEAN, scale=POP_STD, size=SAMPLE_SIZE)

# --- 3. Calculate Sample Statistics and T-Interval (Standard Practice for CI) ---

sample_mean = np.mean(sample_data)
# Use ddof=1 for sample standard deviation (unbiased estimator)
sample_std = np.std(sample_data, ddof=1)
df = SAMPLE_SIZE - 1 # Degrees of Freedom

# Calculate Standard Error (SE)
standard_error = sample_std / np.sqrt(SAMPLE_SIZE)

# Calculate the critical T-value for a two-tailed test
# ppf is the percent point function (inverse of CDF)
t_critical = t.ppf(1 - alpha/2, df)

# Calculate Margin of Error (ME)
margin_of_error = t_critical * standard_error

# Calculate Confidence Interval (CI = Sample Mean ± ME)
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# --- 4. Print Results (CI) ---
print("--- Confidence Interval Calculation ---")
print(f"Sample Size (n): {SAMPLE_SIZE}")
print(f"Sample Mean (x̄): {sample_mean:.4f}")
print(f"Sample Standard Deviation (s): {sample_std:.4f}")
print(f"Degrees of Freedom (df): {df}")
print(f"Confidence Level: {CONFIDENCE_LEVEL*100}%")
print(f"Critical T-value: {t_critical:.4f}")
print(f"Margin of Error: {margin_of_error:.4f}")
print(f"95% Confidence Interval: ({ci_lower:.4f}, {ci_upper:.4f})")
print(f"Interpretation: We are 95% confident that the true population mean lies between {ci_lower:.4f} and {ci_upper:.4f}.")
print("---------------------------------------\n")


# --- 5. Plotting the Data and CI (Figure 1) ---

plt.figure(figsize=(12, 7))

# Plot the histogram of the sampled data
plt.hist(sample_data, bins=15, density=True, color='#2ecc71', edgecolor='black', alpha=0.7, label='Sample Data Histogram')

# Plot a fitted normal distribution curve for visual comparison
x = np.linspace(sample_data.min(), sample_data.max(), 100)
# Use the sample mean and sample standard deviation to fit the curve
p = norm.pdf(x, sample_mean, sample_std)
plt.plot(x, p, 'k', linewidth=2, label='Fitted Normal PDF')

# Add Mean Line
plt.axvline(sample_mean, color='red', linestyle='--', linewidth=2, label=f'Sample Mean ({sample_mean:.2f})')

# Add Confidence Interval Lines
# Lower CI bound
plt.axvline(ci_lower, color='blue', linestyle='-', linewidth=3, label=f'{CONFIDENCE_LEVEL*100}% CI Bound')
# Upper CI bound
plt.axvline(ci_upper, color='blue', linestyle='-', linewidth=3)
# Shade the Confidence Interval region
plt.axvspan(ci_lower, ci_upper, color='blue', alpha=0.1, label='Confidence Interval Region')


# Add Plot Labels and Title
plt.title(f'Figure 1: Sampled Data and {CONFIDENCE_LEVEL*100}% Confidence Interval', fontsize=16)
plt.xlabel('Data Values', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.legend()
plt.grid(axis='y', alpha=0.4)


# --- 6. Z-Score Calculation and Plotting (Figure 2) ---

z_scores = calculate_z_scores(sample_data)

# Print Z-score statistics (should be close to 0 and 1)
z_mean = np.mean(z_scores)
z_std = np.std(z_scores, ddof=1)
print(f"--- Z-Score Standardization Results ---")
print(f"Z-Score Mean (must be ≈ 0): {z_mean:.4f}")
print(f"Z-Score Standard Deviation (must be ≈ 1): {z_std:.4f}")
print("---------------------------------------\n")


plt.figure(figsize=(12, 7))

# Plot the histogram of Z-scores
plt.hist(z_scores, bins=15, density=True, color='#8e44ad', edgecolor='black', alpha=0.7, label='Z-Score Histogram')

# Plot the theoretical Standard Normal Distribution curve
x_z = np.linspace(z_scores.min(), z_scores.max(), 100)
p_z = norm.pdf(x_z, 0, 1) # Standard Normal (Mean=0, StdDev=1)
plt.plot(x_z, p_z, 'k', linewidth=2, linestyle='--', label='Standard Normal PDF ($\mu=0, \sigma=1$)')

# Add Mean Line (at 0)
plt.axvline(0, color='red', linestyle='-', linewidth=2, label='Z-Score Mean (0)')

# Add one standard deviation lines (at -1 and +1)
plt.axvline(-1, color='gray', linestyle=':', linewidth=1.5, label='$\pm 1$ Standard Deviation')
plt.axvline(1, color='gray', linestyle=':', linewidth=1.5)

# Add Plot Labels and Title
plt.title('Figure 2: Distribution of Z-Scores (Standardized Data)', fontsize=16)
plt.xlabel('Z-Score (Standard Deviations from the Mean)', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.legend()
plt.grid(axis='y', alpha=0.4)

# Display both figures
plt.show()
```
